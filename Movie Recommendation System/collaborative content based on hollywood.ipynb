```python
import ast
def convert(obj):     ##this obj is a string.we are creating a helper fucntion to convert genre column into desired format.
    List = []                       ## a list to store this list temporarily
    for i in ast.literal_eval(obj): ##after this the obj will become an integer.For loop 
        List.append(i['name'])      ## appending this the list taking id name and appending it into the list we created
    return List                     ## returning the new list we created
```

## Collaborative Based Filtering : 

```python
movies_ratings.head()

## we have userID,movieID,ratings columns which tells us which user has rated this movie this ratings for example userid = 1 has 
## rated movieid = 2 rating =3. and then the same user has rated movie with id = 5 ratings = 10.
```

```python
movies_ratings.shape
```

```python
x = movies_ratings.groupby('userId').count()['rating'] > 150
filtered_users = x[x].index

## we will keep those users only who have given ratings more than 150 times to keep good results otherwise user who have given 
## ratings to only one movie and they have given 10 ratings to that movie will come up we dont want that. 
```

```python
filtered_rating = movies_ratings[movies_ratings['userId'].isin(filtered_users)]
```

```python
y = filtered_rating.groupby('title').count()['rating'] >= 10
famous_movies = y[y].index
```

```python
famous_movies
```

```python
final_ratings = filtered_rating[filtered_rating['title'].isin(famous_movies)]
```

```python
pt = final_ratings.pivot_table(index='title',columns='userId',values='rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt.head()
```

```python
## we have created a pivot table of users and movies that which user has rated which movie how many ratings we used only those 
## user which have rated over 200 movies and which movies they have rated
```

```python
## every movie here is a vector
## we will calculate distance of every vector from one another
## for this we will use Cosine distance
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
def recommend(movie_name):
    #from movie name fetch index
    index = np.where(pt.index==movie_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    
    for i in similar_items:
        print(pt.index[i[0]])    
```

```python
recommend('Batman Begins')
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags'].apply(stem)
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')

## we will be converting our tags into vectors. for that we need to use countvectorizer from sklearn
## we are making cv and we are taking max features to 5000 meaning the amount of words to form a vector and 
## stop_words='english' means we are taking stop words from english lang or dictionary
```

```python
cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray() ## we are creating a variable vector and strong our array 
```

```python
## we will be using cosine distance instead of euclidean distance becasue we have a large dataset
## we will be calculating angle the smaller the angle the similar the movies
## distance is inversley propotional to similarity
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors) ## we are passing our vectors into cosine similarity
```

```python
cosine_similarity(vectors).shape 
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key = lambda x:x[1])[1:6]
```

```python
def recommendcon(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommendcon("Toy Story")
```
